From the dataset we have we want to optimize the columns to display their apprpriate values.


For starters, we begin with the crew column where we are only interested with the Director. Lets write some code to filter out the Director from the crew and generate their own column.

```python
# begin by writing a function to extract the director's name from the crew
import ast
def extract_dir(text):
    Literal = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
          Literal.append(i['name'])
    return Literal


# Apply the function to the crew column
df['crew'] = df['crew'].apply(extract_dir)
# Change the column name to Director
df.rename(columns={'crew': 'director'}, inplace=True)
df['director'].sample(5)
```

Some of the other  columns have values in a json format including other unnecesary information. We have to create a function that converts the JSON string dictionary and then append the name value into an empty list.

ast.literal_eval() is used to convert string to list

```python
def convert(obj):
    Literal = []  # empty list
    for i in ast.literal_eval(obj): # convert string to list
        i['name']   # i is a dictionary
        Literal.append(i['name']) # append the name of the genre
    return Literal

df['genres']=df['genres'].apply(convert)
df.head(5)
```

***Genres***

Now for the most fun part, Genres. Tney entail the categories the movies fall into making our main objective, to visualize the most popular genres over the years and their distributions.


Since they are in a list format for each record, we will have to create a separate dataset for all the genres categories with the respective year releases of the movies they are related to.

* First we will create an empty dictionary with the year as the key and the genres as values
* Loop through each year placing each occurrence of a movie category whether they appear in that year or not
* Count the movie oocurrences with respect to each year and change it to a dataframe


```python
from collections import Counter

genre_counts = {}
for year in df['release_year'].unique():
    yearly_genres = [genre for genres in df[df['release_year'] == year]['genres'] for genre in genres]
    genre_counts[year] = Counter(yearly_genres)

genre_df = pd.DataFrame(genre_counts).T.fillna(0)

genre_df.head()
```

Now that the dataset has been created, we can now get into visualizing it.

But first we need to calculate the total counts to see the overall popularity of the genres.

```python
genre_df.sum().sort_values(ascending=False)
```

We can plot this:

```python
# we can have the genres on the x axis which in this case are the index arranged in decreasing order
#total count values on the y axis similarly arranged in decreasing order.
plt.figure(figsize=(12, 6))
sns.barplot(x=genre_df.sum().sort_values(ascending=False).index, y=genre_df.sum().sort_values(ascending=False).values, color='green')
plt.xlabel('Genre')
plt.ylabel('Total Count')
plt.title('Total Count of Movies by Genre')
plt.xticks(rotation=90)
plt.show()
```

We can now plot the number of movies for each genre over time using a line chart to see how the popularity of each genre has changed.

```python
plt.figure(figsize=(12, 6))
for genre in genre_df.columns:
  plt.plot(genre_df.index, genre_df[genre], label=genre)
  plt.xlabel('Year')
  plt.ylabel('Number of Movies')
  plt.title('Popularity of Genres Over Time')
  plt.legend()
```

Seeing it be alittle bit crowded, we can try to filter out and see what the distribution for a specific genre, lets say, action, across the years will be.

```python
# Filter the genre_df for the 'Action' genre
action_df = genre_df['Action']

# Plot the number of action movies over time
plt.figure(figsize=(12, 6))
plt.bar(action_df.index, action_df.values, label='Action', width=0.8, color='red')
plt.xlabel('Year')
plt.ylabel('Number of Action Movies')
plt.title('Popularity of Action Movies Over Time')
plt.legend()
plt.show()
```

This looks better!
